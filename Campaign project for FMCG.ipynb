```python
import numpy as np
import matplotlib.pyplot as plt

# Sad dataset (min manat il)
before_campaign = np.array([1200, 1500, 1300, 1400, 1600, 1700,1900])
after_campaign = np.array([1800, 2000, 2200, 2400, 2500, 2700,3000])
months = ["Yan", "Fev", "Mar", "Apr", "May", "yn","Iyul"]

# art
plt.figure(figsize=(8,5))
plt.plot(months, before_campaign, marker='o', label="Kampaniya nc")
plt.plot(months, after_campaign, marker='o', label="Kampaniya sonra")
plt.title("Satlarn Kampaniyadan vvl v Sonra Mqayissi", fontsize=14)
plt.xlabel("Aylar")
plt.ylabel("Satlar (AZN)")
plt.legend()
plt.grid(True)
plt.show()
```
